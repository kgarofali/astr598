# Algorithm Description: Mergesort
We want to merge two sorted arrays. Imagine we have two different sorted arrays, array a[] and array b[] and we want to combine them in the final result, which we'll call array c[]. 

The first thing we do is compare a[0] and b[0] and then put in the lowest. If b[0] is lower, you then compare b[1] and a[0]. If b[1] is still lower, then you next compare b[2] and a[0], and so on. You increment the index of array c[] by one each time. Once one of the arrays has been completely depleted, you no longer need to compare, rather you copy the remainder of the other sorted array into array c[]. 

Mergesort uses divide & conquer, recursion and is a stable sort. 


# Algorithm Code: Mergesort 

```
# assume a.length + b.length = c.length 
void merge(int a[], int a.length, int b[], int b.length, int c[], int c.length){
    int k = 0; 
    int i = 0;
    int j = 0;
    while(k < c.length){
        while((i < a.length) && (j < b.length)){
            if(a[i] < b[j]){
                c[k] = a[i];
                i++;
                k++;
            }
            #takes care of equal to case
            else{
                c[k] = b[j];
                j++;
                k++;
            }
            
        }
    }
    
    #only one of below two loops will run
    #because we have already exited the other while loop
    #so only the leftover array will be copied over 
    while(i < a.length){
        c[k] = a[i];
        i++;
        k++;
    }
    while(j < b.length){
        c[k] = b[j];
        j++; 
        k++;
    }



}
```

What about the special case for with both arrays are of length one? For a single element array, you don't need to do any sortinh. When you combine them with merge, you get a sorted 2-element array. So the divide & conquer approach is to divide your large array down into 2 single-element arrays, and then conquer via the merge approach.
