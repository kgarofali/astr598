#Linked Lists
Linked lists are recursive data structures (i.e. linkedlist <--> node + linkedlist). They are also dynamic. 

Basic structure for linked list, w/ example method: 


`` class LinkedList {
    Node head;
    Node tail;
    }
    head = null;
    tail = null;
    LinkedList L;
    L=new LinkedList();
    L.insert_at_head(4)
``


Contain nodes--each node has data and next. Basic structure for constructing node: 

`` class Node { 
        int data;
        Node next; #you can reference class in class itself, in C or C++ would need a pointer
        }
``

There are three basic cases for linked list: empty linked list (head and tail null, case 1), lots of nodes (tail null, case2), or one node (head and tail equal, case3). Node values for many nodes (case2): 

```
  Node n1; #head node 
  Node n2; 
  Node n3; #tail node 
  n1.data = 1;
  n1.next = n2;
  n2.data = 2;
  n2.next = n3;
  n3.data = 3;
  n3.next = null;
 
```

Traversing a linked list in a loop: 

```
Node n;
n = n1;
while(n != null){
    system.out.println(n.data);
    n = n.next
}
```

##Insert at Tail

We don't want to to delete at tail for single-linked list because we would have to traverse whole linked list, which is no longer O(1). Inserting at tail is like dealing with a queue--you may want to delete at head and insert at tail for a queue. 

```
insert_at_tail(i){
    Node n = new Node();
    n.data = i;
    n.next = null;
    #empty case is special (head == null means tail is also null)
    if(head == null){
        head = n;
        tail = n;
    }
    #non-empty case, one node 
    else if(head == tail){
        tail.next = n; 
        tail = n;
        head.next = n; #good to check in one node case 
    }
    #non-empty case, many nodes, don't need to worry about head 
    else{
        tail.next = n;
        tail = n;
   }
}
```
