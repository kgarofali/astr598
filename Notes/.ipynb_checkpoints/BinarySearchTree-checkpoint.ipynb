# Binary Search Tree

We have already discussed binary heaps, and now we'll move on to binary search trees (both are binary trees). There are also kd-trees, which are space partitioning trees. 

Condition for binary search tree: every node on left sub-tree is less than parent, and every node on the right sub-tree is greater than the parent. 

Three types of traverse: pre_order, in_order, and post_order. 

## BST: Code

Implemented w/ binary recursion. 

```


class Node{
    int data;
    Node left;
    Node right;
}


class BST{
    Node root;
    int N;
    
    public BST(){
        root = null;
        N = 0;
    }
    
    public void insert(int i){
        Node n;
        n = new Node();
        n.data = i;
        n.left = null;
        n.right = null;
        N++; 
        
        if(root == null){
            root = n;
        }
        else{
            local_insert(root,n);
        }
    }
    
    public void local_insert(Node local_root, Node n){
        #n is new Node to insert locally  
        if(local_root == null){
            system.out.println("ERROR")
            exit 
        }
        #if this is true, go left 
        if(n.data < local_root.data){
            #if nothing there, insert 
            if(local_root.left == null){
                local_root.left = n;
            }
            #if there is something there, then
            #use insert (using recursion)
            else{
                insert_local(local_root.left, n)
            }
            
        }
        else{
            if(local_root.right == null){
               local_root.right = n;
            }
            else{
                insert_local(local_root.right, n)
            }
                
    }
    
    
    
    public boolean find(int i){
        return local_find(root, i);
    }
    
    public boolean local_find(Node local_root, int i){
        #base case one 
        if(local_root == null){
            return False
        }
        #base case two 
        else if(local_root.data == i){
            return True
        }
        else if(i < local_root.data){
            return local_find(local_root.left,i)
        }
        else if(i > local_root.data){
            return local_find(local_root.right, i)
        }
    }
    
    public void traverse_pre_order(){
        local_traverse_pre_order(root);
    }
    public void local_traverse_pre_order(Node local_root){
        if(local_root.data == null){
            return;
        }
        else{
            system.out.println(local_root.data)
            local_traverse_pre_order(local_root.left)
            local_traverse_pre_order(local_root.right)
        }
    }    
    
    public void traverse_in_order(){
        local_traverse_in_order(root);
    }
    public void local_traverse_in_order(Node local_root){
        if(local_root.data == null){
            return;
        }
        else{
            local_traverse_in_order(local_root.left)
            system.out.println(local_root.data)
            local_traverse_in_order(local_root.right)
            
        }
    }    
    
    public void traverse_post_order(){
        local_traverse_post_order(root);
    }
    public void local_traverse_post_order(Node local_root){
        if(local_root.data == null){
            return;
        }
        else{
            local_traverse_post_order(local_root.left)
            local_traverse_post_order(local_root.right)
            system.out.println(local_root.data)
            
        }
    } 
    
    public int min(){
       Node local_root;
       int i;
       if(root == null){
           system.out.println("ERROR")
           exit
       }
       local_root = root;
       i = local_root.data;
       while(local_root != null){
           i = local_root.data;
           local_root = local_root.left;
       }
       return i;
    }
    
    public int max(){
       Node local_root;
       int i;
       if(root == null){
           system.out.println("ERROR")
           exit
       }
       local_root = root;
       i = local_root.data;
       while(local_root != null){
           i = local_root.data;
           local_root = local_root.right;
       }
       return i;
    }    
    
```
