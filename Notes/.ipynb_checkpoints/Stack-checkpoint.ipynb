## Stack Implementation w/ Linked List 

```
class Stack{
    private Node head; 
    private N; #size of stack 
    
    #insert_at_head method 
    #this could fail if you run out of memory
    push(int i){
        Node n = new Node();
        n.data = i;
        n.next = head;
        head = n; 
        N++ # N = N + 1 
    }
    
    int pop(){
        int i;
        if(head == null){
            system.out.println("ERROR");
            system.exit(1);
        }
        i = head.data;
        head = head.next; # in C or C++ you would need to then free up head node
        N = N - 1; 
        return i;
    }
    int top(){
        int i; 
        i = head.data;
        return i;
    }
    int size(){
        return N;
    }
    boolean isEmpty(){
        if(head == null){
            return True;
        }
        else{
            return False
        }
    }
}



```

## Stack Implementation w/ Array 

With a fixed size array we have a max number of elements. With this implementation we have the problem of the stack being full. 

```
class Stack{

    private int top = -1;
    private a int a[];
    a = new int[MAX];
    
    #this will fail if you're already at max in C and C++
    #things like java will check boundaries and throw errors
    void push(int i){
        if(top == max){
            system.out.println("ERROR");
            system.out.exit(1);
        }
        top++;
        a[top] = i;
        N++
    }
    
    #need error checking b/c it's more likely to run out of space
    int pop(){
        int i;
        if(N == 0){
            system.out.println("ERROR");
            system.out.exit(1);
        }
        i = a[top];
        top--;
        N--;
    
    }
    
}   
 
 
 
```

```

class Queue{
    private Node head;
    private Node tail;
    private int N;
    public Queue(){
        head = null;
        tail =null;
        N = 0;
    }
    
    public put(int i){
        Node n = new Node();
        n.data = i; 
        n.next = null;
        #null linked list
        if(head == null){
             head = n;
             tail = n;
        }
        #one node linked list
        else if(head == tail){
            head.next = n;
            tail = n;
        }
        else{
            tail.next = n;
            tail = n;
        }
        N++
    }
    
    int get(){
        if(head == null){
            ERROR
        }
        i = head.data;
        head = head.next;
        N = N - 1;
        return i;
    }

    int front(){
        if(head == null){
            ERROR
        }
        i = head.data;
        return i;
    }
    
    
}
```
