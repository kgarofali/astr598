#### Example of stable sort from last time: 

Suppose you have a list of cities (unsorted) but sorted on time. You can use a stable sort to sort the cities, while keeping the times still sorted (think departure time displays at airports). **mergesort** is a stable sort, **heapsort** is not a stable sort. 


# mergesort code

Last time: two arrays, a[] and b[] that we want to sort into resulting array c[], with indices i, j and k. Use the idea of recursion and divide & conquer: for large array split into smaller pieces and then merge them. Now we are assuming that we have some array a[], which we divide into parts (old a[], old b[]) and we allocate space to merge these sorted parts in aux (old c[]). O(NlogN) algorithm. 

```

#could call this something mergesort(int a[]) as well 
public static sort(int a[]){
    
    #allocate auxiliary array space 
    aux = new int[a.length];
    
    #function/method overloading: not the same
    #as above sort, b/c it takes diff params
    #signature of function is name of function
    #and its input params and types. these two 
    #functions have diff signatures b/c params 
    #are not the same 
    sort(a,0,a.length-1);
}

private static void sort(int a[], int lo, int hi){
    #just want to sort from lo to hi values 
    #in recursion, we'll call this same function
    #from within the function, but with diff params 
    #in theory, if you called it with the same params
    #it would be like a loop. for recursion you need
    #a base case, and it must become smaller with each step
    
    int mid;
    #base case of recursion 
    #only one element, so just return 
    if(hi <= lo) return;
    mid = lo + (hi - lo) / 2; 
    sort(a, lo, mid); #sort left-half
    sort(a, mid+1, hi); #sort right-half 
    merge(a,lo,mid,hi);
    
}

merge(int a[], int lo, int mid, int hi){
    
    int i = lo;
    int j = mid + 1;
    int k;
    #first make a copy of input array 
    #in the given range 
    for(k=lo;k<hi;k++){
        aux[k] = a[k];
    }
    #now we want to merge sorted parts 
    #into one array 
    k = lo;
    while(k<=hi){
        #both arrays are not finished 
        #only exits if i==mid or j==hi
        while((i<=mid) && (j<=hi)){
            if(aux[j] < aux[i]){
                a[k] = aux[j];
                j++;
                k++;
            }
            else{
                a[k] = aux[i];
                i++;
                k++;
           }
        }
        #either i>mid or j>hi
        #so only one of below loops will run
        while(i<=mid){
            a[k] = aux[i];
            k++;
            i++;
        }
        while(j <= hi){
            a[k] = aux[j];
            j++;
            k++;
        }
    }

}

```
